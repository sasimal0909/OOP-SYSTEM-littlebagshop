```python
def summary(df, col):
  _min = df[col].min()
  _max = df[col].max()
  _mean = df[col].mean()
  _std = df[col].std()
  _median = df[col].median()

  print(
      f"Min of the data {_min}",
      f"Max of the data {_max}",
      f"Mean of the data {_mean}",
      f"Std of the data {_std}",
      f"Median of the data {_median}",
      sep="\n"
  )
```
